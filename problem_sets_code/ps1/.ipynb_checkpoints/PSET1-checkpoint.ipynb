***a) Chứng minh rằng các đường thẳng song song trong hệ quy chiếu thế giới vẫn song song trong hệ quy chiếu camera. [4 điểm]***

Gọi $P_1$ là một điểm thuộc vaof đường thẳng $l_1$ có vecto chỉ phương $u$, $P_2$ là một điêm thuộc vào $l_2$

Khi đó, ta biểu diễn
$x_1 = P1 + t_1 u$
$x_2 = P2 + t_2 u$

Điểm $x_1$ và $x_2$ trong hệ tọa độ camera là:
$x_1' = R x_1 + t = R(P1 + t_1 u) + t = RP1 + R t_1 u + t = P_1' + t_1 R u$
$x_2' = P_2' + t_2 R u$

Vậy $x_1'$ và $x_2'$ vẫn song song trong hệ tọa độ camera (cùng vecto chỉ phương $R u$)

***Xét một hình vuông đơn vị pqrs trong hệ quy chiếu thế giới, trong đó p, q, r và s là các điểm. Hỏi hình vuông đó trong hệ quy chiếu camera có luôn có diện tích đơn vị không? Chứng minh hoặc đưa ra phản ví dụ. [4 điểm]***

Diện tích của một hình bình hành được tính như nào?

Yah, chính là $S = \| pq \times ps \|$

Theo giả thiết, ta có $\| pq \times ps \| = 1$

Ta có $p'q' = R pq$, $p'r' = R ps$ (t đã bị triệu tiêu khi trừ hai điểm nhoa)

Ta cùng đi tính S' thôi:

\begin{align*}
    S' &= \| p'q' \times p's' \| \\
       &= \| (R pq) \times (R ps) \\
       &= \| R (pq \times ps) \| \\
       &= \| pq \times ps \| \text{( Vì $det(R) = \pm  1$} \\
       &= 1 (điều phải chứng minh)
\end{align*}


***c) Bây giờ chúng ta hãy xem xét các phép biến đổi affine, là bất kỳ phép biến đổi nào bảo toàn tính song song. Các phép biến đổi affine bao gồm không chỉ phép quay và phép tịnh tiến, mà còn cả phép co giãn và phép trượt. Cho một vectơ p nào đó, một phép biến đổi affine được định nghĩa là:***

        $A(p) = Mp + b$

***trong đó M là một ma trận khả nghịch. Chứng minh rằng dưới bất kỳ phép biến đổi affine nào, tỷ lệ của các đoạn thẳng song song là bất biến, nhưng tỷ lệ của các đoạn thẳng không song song thì không bất biến. [6 điểm]***

Giải:

Giả sử $l_1 \parallel l_2$ có vecto chỉ phương là $u$

Xét $A$, $B$ thuộc vào đường thẳng $l_1$

Xét $C$, $D$ thuộc vào đường thẳng $l_2$

Ta có tỷ lệ:

\begin{align*}
    \frac{\|AB\|}{\|CD\|} = \frac{t_{BA} \|u\|}{t_{DC} \|u\|} = \frac{t_{BA}}{t_{DC}}
\end{align*}

Sử dụng câu a), ta có:

\begin{align*}
    A'B' = t_B M u - t_A M u = t_{BA} M u = t_{BA} v \\
    C'D' = t_D M u - t_C M u = t_{DC} M u = t_{DC} v
\end{align*}

Ta xét tỷ lệ:

\begin{align*}
    \frac{\|A'B'\|}{\|C'D'\|} = \frac{\|t_{BA} v\|}{\|t_{DC} v\|} = \frac{t_{BA}}{t_{DC}} = \frac{\|AB\|}{\|CD\|}
\end{align*}

Với $l_1 \nparallel l_2$

Xét tương tự:

\begin{align*}
    \frac{\|AB\|}{\|CD\|} = \frac{t_{BA} \|u\|}{t_{DC} \|v\|} \\
    \frac{\|A'B'\|}{\|C'D'\|} = \frac{t_{BA} \|u'\|}{t_{DC} \|v'\|} \neq \frac{\|AB\|}{\|CD\|}
\end{align*}

Điểm giao nhau của hai đường thẳng được tính theo công thức:

***b) Sử dụng ba điểm tiêu, chúng ta có thể tính toán ma trận nội của camera được sử dụng để chụp ảnh. Thực hiện điều này trong hàm compute_K_from_vanishing_points(). [10 điểm]***

Về kiến thức phần này thì cũng đã được học trong note rồi, nhắc lại xíu nhé!

Gọi d là tập hợp các vecto chỉ phương, v là các điểm tiêu.

Ta có $v = K d$

Với các giải thuyết đề bài thì ma trận nội tại của camera lúc này sẽ là:

$$
K = \begin{bmatrix}
    f & 0 & c_x \\
    0 & f & c_y \\
    0 & 0 & 1
\end{bmatrix}
$$

Nhưng với giả sử này, chúng ta vẫn cần 3 giả thuyết nữa (còn 3 bậc tự do trong ma trận K). Mà 3 điểm tiêu mà đề cho chúng ta không cung cấp dữ liệu về d (chỉ cung cấp v). Nên chúng ta sẽ đặt các giả sử giống trong note như sau: Giả sử các điểm tiêu này nằm trên 3 mặt phẳng vuông góc nhau.

Khi đó, ta vận dụng công thức $cos(d1, d2) = \frac{v_1^T w v_2}{\sqrt{v_1^T w v_1} \sqrt{v_2^T w v_2}} = 0$

Với điểm $v1, v2$ là điểm tiêu của đường thẳng $d1, d2$

Ta sẽ có 3 phương trình:
\begin{align*}
    v_1^T w v_2 = 0 \\
    v_2^T w v_3 = 0 \\
    v_3^T w v_1 = 0
\end{align*}

Với ma trận $w$ là:
$$
w = \begin{bmatrix}
    w_1 & 0 & w_4 \\
    0 & w_1 & w_5 \\
    w_4 & w_5 & w_6
\end{bmatrix}
$$

Ta sẽ tìm ma trận w nhé!

Với $v_1^T w v_2 = 0$, ta có:

\begin{align*}
    \begin{vmatrix}
        v_{10} & v_{11} & v_{12}
    \end{vmatrix} \begin{bmatrix}
        w_1 & 0 & w_4 \\
    0 & w_1 & w_5 \\
    w_4 & w_5 & w_6
    \end{bmatrix} \begin{bmatrix}
        v_{20} \\ v_{21} \\ v_{22}
    \end{bmatrix} = 0 \\
    \begin{bmatrix}
        v_{10} w_1 + v_{12} w_4 &
        v_{11} w_1 + v_{12} w_5 &
        v_{10} w_4 + v_{11} w_5 + v_{12} w_6
    \end{bmatrix} \begin{bmatrix}
        v_{20} \\ v_{21} \\ v_{22}
    \end{bmatrix} = 0 \\
    w_1 (v_{10} v_{20} + v_{11} v_{21}) + w_4 (v_{12} v_{20} + v_{10} v_{22}) + w_5 (v_{12} v_{21} + v_{11} v_{22}) + w_6 (v_{12} v_{22}) = 0 \\
    \begin{bmatrix}
        v_{10} v_{20} + v_{11} v_{21} &
        v_{12} v_{20} + v_{10} v_{22} &
        v_{12} v_{21} + v_{11} v_{22} &
        v_{12} v_{22}
    \end{bmatrix} \begin{bmatrix}
        w_1 \\ w_4 \\ w_5 \\ w_6
    \end{bmatrix} = 0
\end{align*}

Giờ chỉ việc mô hình và tìm w bằng SVD nữa là chúng ta sẽ có $\omega$. Sau đí sử dụng Cholesky Decomposition để tìm ma trận $K^T$ và chuẩn hóa lại nữa là sẽ xong câu này rồi!
